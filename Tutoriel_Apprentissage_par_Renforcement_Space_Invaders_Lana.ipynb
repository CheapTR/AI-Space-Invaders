# Pourquoi des IA?

Pouvoir faire en sorte de faire une tâche automatiquement permet de la réaliser plus rapidement et permet également aux humains de se concentrer sur des tâches plus complexes. Par exemple, lors de la construction de voitures, le tout est automatisé afin de les produire à la chaîne. Automatiser des problèmes complèxes qui changent demande à ce que le programme utilisé s'adapte à la tâche devant lui, donc qu'il apprenne.

## Exemple de problème qui bénéficie d'une IA

Par exemple, un problème connu en computation est celui du *travelling salesman*, où il faut aller à différents points sur une carte en le moins de temps (distance) possible. Ce problème devient exponentiellement plus difficile à mesure que l'on ajoute des points sur la carte. La solution simple, mais longue à ce problème est de tester toutes les possibilités possibles et de prendre la plus optimale. Ceci est long pour un humain mais l'est moins pour un ordinateur. Cependant, l'ordinateur peut prendre beaucoup de temps aussi à faire ceci. C'est pourquoi une solution acceptée dans beaucoup de cas n'est pas nécessairement la meilleure, mais bien plutôt une qui est acceptable et qui n'est pas trop loin de celle qui est la meilleure, car elle prendrait trop de temps à trouver. Le but serait donc de faire un programme qui trouve une solution acceptable rapidement. Un programme qui apprend à reconnaitre les façons de procéder permettrait donc d'accéder à cette solution acceptable plus rapidement que si il procédait de façon aléatoire, par exemple.

## Des points négatifs?

Il faut retenir qu'une IA ne va pas forcément trouver la solution optimale à certains problèmes, mais bien trouver une solution acceptable. Il faut également parfois la superviser, puisqu'elle n'est pas infaillible, tout comme les humains. Un exemple d'IA qui ne trouve pas de solution optimale peut se trouver dans le jeu *Super Mario Bros.* avec son tout premier niveau. Vous vous souvenez sans doute du passage secret dans un tuyau dans ce niveau. L'IA ne sachant pas qu'il éxiste des passages secrets, va simplement ignorer le tuyau et le considérer comme les autres plateformes du niveau. Un humain peut simplement se faire dire "si tu appuies par en bas sur ce tuyau, il y a un passage secret", alors que pour une IA dire celà est plus compliqué et nécéssite du code plus spécifique, alors que l'on veut qu'il s'adapte. L'IA va cependant trouver éventuellement que d'appuyer sur droite et courir en même temps en tout temps va mener vers un résultat qui sera très rapide, malgré le fait qu'elle ne trouvera probablement jamais le passage secret à moins de lui dire explicitement.

L'IA est très demandante pour un ordinateur lors de son entraînement, et peut donc prendre beaucoup de temps!

Pour résumer: Les IA servent à automatiser des tâches complexes qui peuvent changer. Il arrive assez fréquemment que l'IA n'arrive pas à la solution la plus optimale si celle-ci prendrait trop de temps à aboutir, et se contentera d'un résultat acceptable. Elle est également fiable, mais demande un peu de supervision.


# Quels sont les concepts derrière l'apprentissage par renforcement exactement?

Ces prochains segments serviront à définir plusieurs termes que nous utiliserons tout au long du tutoriel.

## Environnement

L'environnement est ce avec quoi l'agent va interagir. Dans le cas du jeu *Super Mario Bros.* , ce serait donc le niveau, incluant les plateformes, les ennemis et l'arrière plan. (l'interaction pour l'arrière plan se limitant à l'agent qui le voit, l'agent doit quand même savoir que l'arrière plan n'a pas à être pris en compte)

## Agent

L'agent est un élément qui ira intéragir avec l'environnement. Il peut être humain ou non et il n'y a pas de limite à combien il peut en avoir dans un environnement. Dans le cas du jeu *Super Mario Bros.* , ce serait donc le(s) joueur(s). Pour un agent, les autres agents sont vus comme faisant partie de l'environnement.

## Entraînement

L'entraînement est le processus d'introduire un agent dans un environnement pour qu'il interagisse avec. Dans le cas du jeu *Super Mario Bros.* , ce serait donc le temps de jeu.

## Récompense et pénalité

Une récompense ou une pénalité est donnée à l'agent pour ses actions afin de lui faire comprendre les actions qui sont favorables ou non à son but. Dans le cas du jeu *Super Mario Bros.* , les récompenses dans l'ordre des priorités seraient donc faire du progrès dans le niveau (plus la distance vers le drapeau est basse, plus l'on a progressé), le temps restant lorsque l'on termine le niveau, ainsi que le score. Les pénalités dans l'ordre des priorités seraient de se faire toucher par un ennemi, de ne plus avoir de temps, ainsi que de perdre du progrès dans le niveau.

## État

L'état de l'environnement est les situations que l'agent va rencontrer et qui vont faire réagir l'agent lors de son entraînement. Dans le cas du jeu *Super Mario Bros.* , ce serait donc le placement de tous les pixels à l'écran.

## Action

Les réactions de l'agent face à l'état de l'environnement sont les actions, qui peuvent lui être favorables, défavorables, ou neutres. Le but d'un agent est de faire des actions qui vont lui permettre de maximiser ses récompenses au travers de son entraînement. Dans le cas du jeu *Super Mario Bros.* , ce seraient les touches sur lesquelles le joueur appuie ou n'appuie pas.

## Stratégie

La stratégie de l'agent est le plan qu'il se fait au fil de son entraînement. Sans être l'agent, il faut soit se la faire expliquer par l'agent, soit interpréter ses actions pour déduire sa stratégie. Dans le cas du jeu *Super Mario Bros.* , ce seraient les pensées du joueur du genre "Je sais qu'il y a un passage secret dans ce tuyau, je vais éviter cet ennemi, je ne sais pas ce qui vient après dans le niveau, je vais donc procéder plus lentement, etc."

## Espace d'état

L'espace d'état est le regroupement de tous les états possibles. Dans le cas du jeu *Super Mario Bros.* , ce serait les combinaisons de la position du joueur, la position des ennemis, la position des plateformes, la position des pièces, etc. possibles.

## Espace d'actions

L'espace d'actions est le regroupement de toutes les actions possibles. Dans le cas du jeu *Super Mario Bros.* , ce serait les possiblités de combinaisons de touches appuyées ou non entre gauche, droite, bas, courir et sauter (haut ne fait rien).

## Processus d'apprentissage de l'IA

L'agent va observer l'environnement, faire des actions, qui seront aléatoires au début. Il va observer les effets de ses actions sur l'environnement et les récompenses qu'il recevra détermineront s'il désire répéter les actions qu'il entreprend ou tenter d'autres actions.

##Q-Learning

Le *Q-Learning* est l'algorithme d'apprentissage le plus utilisé en apprentissage par renforcement. C'est celui qui sera utilisé dans ce tutoriel. Nous le verrons sous le nom de **EpsGreedyQPolicy**.

## Exploration

L'exploration est la phase où l'agent tente de nouvelles possibilités d'actions pour obtenir de meilleures récompenses.

## Exploitation

L'exploitation est la phase où l'agent a trouvé une stratégie pour obtenir des récompenses efficacement et la met en action.

```python
import gym


from gym import logger as gymlogger
from gym.wrappers import Monitor
import glob
import io
import base64
from IPython.display import HTML
from IPython import display as ipythondisplay
from pyvirtualdisplay import Display
display = Display(visible=0, size=(1400, 900))
display.start()
```

```python
def show_video():
  mp4list = glob.glob('video/*.mp4')
  if len(mp4list) > 0:
    mp4 = mp4list[0]
    video = io.open(mp4, 'r+b').read()
    encoded = base64.b64encode(video)
    ipythondisplay.display(HTML(data='''<video alt="test" autoplay 
                loop controls style="height: 400px;">
                <source src="data:video/mp4;base64,{0}" type="video/mp4" />
             </video>'''.format(encoded.decode('ascii'))))
  else: 
    print("Could not find video")
```

```python
def wrap_env(env):
  env = Monitor(env, './video', force=True)
  return env
```

La ligne suivante permet de créer l'environnement avec *Space Invaders*. En python, nous n'avons pas besoin de typer nos variables, donc env est une nouvelle variable à laquelle nous assignons l'environnement de *Space Invaders*.

```python
env = wrap_env(gym.make("SpaceInvaders-v4"))
```

Le bloc suivait fait dix itérations d'un agent qui fait des actions au hasard dans le code.

`episodes = 10` met notre nombre de parties à dix

`for episode in range(1, episodes+1)`: fait une boucle qui initialise une variable **episode** à 1 (gauche de la parenthèse), nous ajoutons +1 à **episodes**, car la boucle ferait neuf itérations, puisque la variable **episode** doit être inférieure à la droite de la parenthèse pour rester dans la boucle. En python, il n'y a pas de {} pour signifier l'inclusion dans un **if**, un **for**, un **while** ou autre. En lieu de ça, c'est avec l'indentation.

`state = env.reset()` Nous initions la variable **state** à un redémarrage de l'environnement. Cette variable sera utilisée pour voir l'état de l'environnement.\
`done = False` Nous initions aussi la variable **done** à faux. Cette variable nous indiquera si le jeu est terminé.\
`score = 0`   Nous initions finalement la variable **score**, qui nous servira à voir la performance de l'agent, à 0.
    
`while not done:` Tant que le jeu n'est pas terminé...\
`env.render()` Montre l'environnement afin que l'on puisse voir les actions que l'agent est en train de faire. Si vous êtes sur Google Colab, vous aurez une erreur à cette ligne.\
`state, reward, done, info = env.step(env.action_space.sample())` La fonction env.step() est l'action que l'agent prendra à la prochaine *frame* (1/60e de seconde). Ce qu'il y a dans les parenthèses, c'est une fonction permettant d'effectuer une action au hasard dans l'espace d'action. Nous voulons avoir l'état de l'environnement (**state**), la récompense (**reward**), si nous avons terminé (**done**), ainsi que les informations de l'environnement (**info**)\
`score += reward` ajoute la récompense de l'action au score de cette itération de jeu

`print('Episode: {}\nScore: {}'.format(episode, score))` Montre quelle itération vient de se terminer et avec quel score

`show_video()` montre la vidéo de ce que l'agent a fait cet épisode-ci

`env.close()` ferme l'environnement


```python
episodes = 10

for episode in range(1, episodes+1):
    state = env.reset()
    done = False
    score = 0
    
    while not done:
        env.render()
        state, reward, done, info = env.step(env.action_space.sample())
        score += reward
    print('Episode: {}\nScore: {}'.format(episode, score))
    show_video()
    
env.close()
show_video()
```

# Début du réseau de neurones

Nous aurons besoin de Numpy, qui permet de faire des mathématiques plus complexes. 

Nous aurons également besoin de Sequential à partir de Tensorflow. Sequential indique que notre réseau de neurones sera une séquence de couches. 

Les couches, que nous importerons également de Tensorflow, seront Dense, signifiant un réseau de neurones entièrement connecté, Flatten, signifiant que le résultat sera un tableau en une dimension qui est plus facile à lire pour un ordinateur, ainsi que Conv2D, qui est la manière d'interpréter les images pour un ordinateur.

L'optimiseur, que nous importerons encore une fois de Tensorflow, sera Adam. Un optimiseur sert à trouver la bonne balance pour un agent.

Le bloc suivant importe tous ces paquets.

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Conv2D
from tensorflow.keras.optimizers import Adam
```

En exécutant la ligne de code suivante, nous pouvons observer l'espace de jeu tel qu'il est donné à l'IA. Il est donné sous le format hauteur, largeur, canaux. La hauteur et la largeur sont données en pixels.

```python
env.observation_space.shape
```

Le bloc de code suivante permet de définir une fonction pour bâtir notre modèle qui prend en compte quatre paramètres: hauteur, largeur, canaux et actions.

Nous voulons que les couches de notre modèle soient en séquentiel.

Nous voulons ajouter une couche Conv2D. Le premier paramètre à l'intérieur de celle-ci sera le nombre de neurones. Plus ce nombre est élevé, plus notre modèle pourra bien comprendre son environnement. Le second et le troisième paramètre à l'intérieur de Conv2D() sont utilisées pour condenser l'image afin que le modèle puisse la comprendre plus facilement. l'activation relu sert à avoir des non-linéarités dans le modèle, ce qui lui servira à apprendre des *patterns* plus complèxes. Le input_shape est la forme de l'écran de jeu. Ceci sera notre entrée.

Nous voulons ajouter une seconde couche Conv2D pour l'entrée. Le nombre de neurones dans celle-ci sera 64.

La couche Flatten écrasera notre entrée dans un tableau à une dimension afin de mieux la lire.

Les deux premières couches Dense servent à avoir plus de neurones dans le réseau, donc plus de paramètres possibles. Il est possible d'en rajouter plus.

La couche Dense finale aura les paramètres actions et une activation linéaire, pour que l'on ait une des actions permises dans *Space Invaders*.

Finalement, nous n'avons qu'à retourner le modèle.


Pour résumer, les couches que nous avons ajoutées seront appliquées l'une après l'autre à chaque *frame* pour que le réseau de neurones puisse se faire ses propres paramètres et choisir ses actions lui-même. Le modèle sera le réseau de neurones de l'agent. Nous pouvons maintenant faire apprendre à notre agent à partir de n'importe quel écran que nous lui présenterons.

```python
def build_model(height, width, channels, actions):
    model = Sequential()
    model.add(Conv2D(32, (8,8), strides=(4,4), activation='relu', input_shape=(3, height, width, channels)))
    model.add(Conv2D(64, (4,4), strides=(2,2), activation='relu'))
    model.add(Flatten())
    model.add(Dense(512, activation='relu'))
    model.add(Dense(256, activation='relu'))
    model.add(Dense(actions, activation='linear'))
    return model
```

Ici, nous définirons les variables que nous utiliserons dans la fonction build_model. La hauteur, la largeur et les canaux sont déjà donnés dans **env.observation_space.shape**. Les actions sont données dans **env.action_space.n**. le .n est le nombre d'actions possibles.

```python
height, width, channels = env.observation_space.shape
actions = env.action_space.n
```

Les prochains blocs sont l'initialisation du modèle que l'agent utilisera pour apprendre de son environnement. Nous effaçons le modèle avant pour nous éviter une erreur dans la fonction **build_agent** que nous ferons plus tard.

```python
del model
```

```python
model = build_model(height, width, channels, actions)
```

# Application de l'apprentissage par renforcement

Maintenant que nous avons le modèle, il est l'heure d'appliquer l'apprentissage par renforcement.

Le bloc suivant est l'importation des paquets que nous aurons besoin pour ce segment.

À partir de **keras-rl2**, nous importerons **DQNAgent**, qui sera notre algorithme d'apprentissage. Il a été développé en 2015 et a été utilisé pour apprendre à jouer à plus de 300 jeux Atari avec le même réseau de neurones. Il nous sera donc bien utile pour *Space Invaders*.

Nous importerons également **SequentialMemory**, qui garde en mémoire les observations, les actions et les récompenses à chaque *frame*. Ce sera les données à partir desquelles notre agent apprendra.

Nous importerons aussi des polices, qui seront le comportement de notre agent. Les polices que nous importerons seront **LinearAnnealedPolicy** ainsi que **EpsGreedyQPolicy**. **EpsGreedyQPolicy** est une police qui incite l'agent à entreprendre l'action qui amènera à la plus grande récompense. **LinearAnnealedPolicy** est une police avec un seuil, nous allons donc lui donner le **EpsGreedyQPolicy** afin que notre agent puisse computer quelle sera l'action à entreprendre.

```python
from rl.agents import DQNAgent
from rl.memory import SequentialMemory
from rl.policy import LinearAnnealedPolicy, EpsGreedyQPolicy
```

Dans le bloc suivant, nous allons bâtir notre agent.

Nous commençons par lui définir les polices que nous avons importées. Nous appliquons la police **EpsGreedyQPolicy** dans la **LinearAnnealedPolicy**, nous insérons l'attribut Epsilon `attr='eps'`, qui sera une probabilité de faire une action au hasard. Cet attribut ne sera pas fixe, il sera diminué au fil de l'entraînement de l'agent. 

Epsilon est donc la probabilité d'être en phase d'exploration à ce moment. **value_max** est la valeur maximale d'Epsilon que nous désirons, qui est à 1 (donc 100%). **value_min** est sa valeur minimale, qui sera à 0.1 (donc 10%). Elle n'est pas à 0, puisque nous désirons avoir un agent qui est capable de perfectionner une solution. Nous voulons tester si la valeur d'Epsilon sera sous 0.2 (donc 20%), alors nous mettons **value_test** à 0.2. L'IA aura des sessions de 10000 étapes, donc nous mettons **nb_steps** à 10000.

Nous allons ensuite définir la mémoire, qui aura une limite de 2000 étapes. Notre agent se souviendra des 2000 dernières étapes qu'il aura faites.

Nous allons ensuite faire notre agent (**dqn**), qui aura notre modèle, notre mémoire et notre police en paramètres. **enable_dueling_network** est une technique que les chercheurs ont conclus qu'elle amenait de meilleurs résultats. Le type de *dueling* sera mis à moyen (**avg** pour *average*). Notre échauffement sera de 1000 étapes.

```python
def build_agent(model, actions):
    policy = LinearAnnealedPolicy(EpsGreedyQPolicy(), attr='eps', value_max=1, value_min=.1, value_test=.2, nb_steps=10000)
    memory = SequentialMemory(limit=2000, window_length=3)
    dqn = DQNAgent(model=model, memory=memory, policy=policy,
                  enable_dueling_network=True, dueling_type='avg',
                  nb_actions=actions, nb_steps_warmup=1000)
    return dqn
```

Nous mettons ensuite notre agent dans une variable. Si vous avez une erreur, retournez à la création du modèle et exécutez les blocs de code.

```python
dqn = build_agent(model, actions)
```

Nous allons ensuite compiler notre agent avec l'optimisateur Adam, et nous aurons un taux d'apprentissage (**lr** pour *learning rate*) de 0.0001. Il est bon de jouer avec le taux d'apprentissage si nous n'avons pas le résultat que nous voulons. Nous pourrions retirer ou ajouter des zéros pour faire des différences significatives.

```python
dqn.compile(Adam(lr=0.0001))
```

Nous allons maintenant passer l'environnement à notre agent pour qu'il puisse faire son entraînement. Nous voulons qu'il fasse 40000 étapes à son entraînement, alors nous mettons **nb_steps** à 40000. Un *step*, soit une étape, est une action de l'agent **visualize** sert à ce que l'on puisse observer l'entraînement de notre agent. Il peut être à *true* (ne fonctionne pas avec Google Colab) ou à *false*, cependant s'il est à *true* ce sera plus long. Nous allons mettre la variable **verbose** à 1. C'est normal que l'entraînement soit très long!
